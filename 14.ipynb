```python
from functools import cache


@cache
def shift_rocks(map_string: str, direction: str):
    map = map_string.split("\n")
    next_map = map.copy()

    if direction == "N":
        for x in range(len(map[0])):
            for y in range(len(map)):
                if map[y][x] == "O":
                    for i in range(y - 1, -2, -1):
                        if i == -1 or next_map[i][x] in "O#":
                            next_map[y] = next_map[y][:x] + "." + next_map[y][x + 1 :]
                            next_map[i + 1] = (
                                next_map[i + 1][:x] + "O" + next_map[i + 1][x + 1 :]
                            )
                            break

        return "\n".join(next_map)

    if direction == "S":
        for x in range(len(map[0])):
            for y in range(len(map) - 1, -1, -1):
                if map[y][x] == "O":
                    for i in range(y + 1, len(map) + 1):
                        if i == len(map) or next_map[i][x] in "O#":
                            next_map[y] = next_map[y][:x] + "." + next_map[y][x + 1 :]
                            next_map[i - 1] = (
                                next_map[i - 1][:x] + "O" + next_map[i - 1][x + 1 :]
                            )
                            break

        return "\n".join(next_map)

    if direction == "E":
        for y in range(len(map)):
            for x in range(len(map[0]) - 1, -1, -1):
                if map[y][x] == "O":
                    for i in range(x + 1, len(map[0]) + 1):
                        if i == len(map[0]) or next_map[y][i] in "O#":
                            next_map[y] = next_map[y][:x] + "." + next_map[y][x + 1 :]
                            next_map[y] = next_map[y][: i - 1] + "O" + next_map[y][i:]
                            break

        return "\n".join(next_map)

    if direction == "W":
        for y in range(len(map)):
            for x in range(len(map[0])):
                if map[y][x] == "O":
                    for i in range(x - 1, -2, -1):
                        if i == -1 or next_map[y][i] in "O#":
                            next_map[y] = next_map[y][:x] + "." + next_map[y][x + 1 :]
                            next_map[y] = (
                                next_map[y][: i + 1] + "O" + next_map[y][i + 2 :]
                            )
                            break

        return "\n".join(next_map)

    raise Exception("Invalid direction")


def get_north_load(map_string: str):
    map = map_string.split("\n")
    total = 0
    for y in range(len(map)):
        for x in range(len(map[0])):
            if map[y][x] == "O":
                total += len(map) - y

    return total
```

```python
test_input = """\
O....#....
O.OO#....#
.....##...
OO.#O....O
.O.....O#.
O.#..O.#.#
..O..#O..O
.......O..
#....###..
#OO..#...."""

assert get_north_load(shift_rocks(test_input, "N")) == 136
```

```python
value = get_north_load(shift_rocks(open("14.txt").read(), "N"))

print(value)
assert value == 111979
```

```python
directions = ["N", "W", "S", "E"] * 3

test_target = """\
.....#....
....#...O#
.....##...
..O#......
.....OOO#.
.O#...O#.#
....O#...O
.......OOO
#...O###.O
#.OOO#...O"""

map = test_input
for direction in directions:
    map = shift_rocks(map, direction)

assert map == test_target
```

```python
directions = ["N", "W", "S", "E"] * 10**2

maps = [test_input]
loop_size = None
loop_start = None
for i, direction in enumerate(directions):
    map = shift_rocks(maps[-1], direction)

    if not loop_size and map in maps:
        loop_size = i - maps.index(map) + 1
        loop_start = maps.index(map)

    maps.append(map)


def get_index(i: int):
    next_index = i % loop_size

    while next_index < loop_start:
        next_index += loop_size

    return next_index


assert maps[get_index(120)] == maps[120] and get_index(120) != 120
```

```python
directions = ["N", "W", "S", "E"] * 10**4

maps = [open("14.txt").read()]
loop_size = None
loop_start = None
for i, direction in enumerate(directions):
    map = shift_rocks(maps[-1], direction)

    if not loop_size and map in maps:
        loop_size = i - maps.index(map) + 1
        loop_start = maps.index(map)
        break

    maps.append(map)


def get_index(i: int):
    next_index = i % loop_size

    while next_index < loop_start:
        next_index += loop_size

    return next_index


total = get_north_load(maps[get_index(4 * 10**9)])
print(total)
```
